# 3. Analysis of values and dependencies in the dataset

```python
# mapping string values to numbers
college_df['type_school'] = college_df['type_school'].map({'Academic': 1,
                                                           'Vocational': 0})

college_df['school_accreditation'] = college_df['school_accreditation'].map({'A': 1,
                                                                             'B': 0})

college_df['gender'] = college_df['gender'].map({'Male': 1,
                                                 'Female': 0})

college_df['interest'] = college_df['interest'].map({'Not Interested': 0,
                                                     'Less Interested': 1,
                                                     'Uncertain': 2,
                                                     'Interested': 3,
                                                     'Very Interested': 4})

college_df['residence'] = college_df['residence'].map({'Urban': 1,
                                                       'Rural': 0})

college_df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

correlation = college_df.corr().round(2)
plt.figure(figsize=(12, 8))
sns.heatmap(data=correlation, annot=True, cmap='coolwarm')
```

# 4. Distribution

```python
r = 6
c = 4

plt.figure(figsize=(13, 13))

for i, column in enumerate(college_df.columns):
    plt.subplot(r, c, i+1)
    dis_dist = sns.distplot(college_df[column], bins=20)
    dis_dist.set_title(f'{column}')  

plt.tight_layout() 
plt.show()
```
