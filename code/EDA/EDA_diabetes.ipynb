# 3. Analysis of values equal 0

```python
# plas_glucose_concentr
zero_count = diabetes_df[diabetes_df['plas_glucose_concentr'] == 0].shape[0]
print(f"Number of values equal to 0 in 'plas_glucose_concentr': {zero_count}")

small_count = diabetes_df[diabetes_df['plas_glucose_concentr'] < 70].shape[0]
print(f"Number of values lower than 70 in 'plas_glucose_concentr': {small_count}")

# blood_pressure
zero_count = diabetes_df[diabetes_df['blood_pressure'] == 0].shape[0]
print(f"Number of values equal to 0 in 'blood_pressure': {zero_count}")

small_count = diabetes_df[diabetes_df['blood_pressure'] < 40].shape[0]
print(f"Number of values lower than 40 in 'blood_pressure': {small_count}")

# skin_thick
zero_count = diabetes_df[diabetes_df['skin_thick'] == 0].shape[0]
print(f"Number of values equal to 0 in 'skin_thick': {zero_count}")

small_count = diabetes_df[diabetes_df['skin_thick'] < 5].shape[0]
print(f"Number of values lower than 5 in 'skin_thick': {small_count}")

# insulin
zero_count = diabetes_df[diabetes_df['insulin'] == 0].shape[0]
print(f"Number of values equal to 0 in 'insulin': {zero_count}")

big_count = diabetes_df[diabetes_df['insulin'] > 60].shape[0]
print(f"Number of values higher than 60 in 'insulin': {big_count}")

# mass_index
zero_count = diabetes_df[diabetes_df['mass_index'] == 0].shape[0]
print(f"Number of values equal to 0 in 'mass_index': {zero_count}")

small_count = diabetes_df[diabetes_df['mass_index'] < 10].shape[0]
print(f"Number of values lower than 10 in 'mass_index': {small_count}")
```

### Analysing rows with multiple 0 values (omitting 'insulin' and 'skin_thick')

```python
# Columns you want to check
columns_to_check = ['plas_glucose_concentr', 'blood_pressure', 'mass_index']

# Create a mask that checks where the values in the selected columns are equal to 0
zero_mask = (diabetes_df[columns_to_check] == 0)

# Count the number of zeros in each row
zero_count_per_row = zero_mask.sum(axis=1)

# Select rows where at least 3 columns have a value equal to 0
rows_with_at_least_3_zeros = diabetes_df[zero_count_per_row >= 2]

# Display the selected rows
print(rows_with_at_least_3_zeros)
```

# 4. Analysis of values and dependencies in the dataset

```python
# mapping string values to numbers
diabetes_df['class'] = diabetes_df['class'].map({'tested_positive': 1,
                                                 'tested_negative': 0})


diabetes_df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

correlation = diabetes_df.corr().round(2)
plt.figure(figsize=(12, 8))
sns.heatmap(data=correlation, annot=True, cmap='coolwarm')
```

# 4. Distribution

```python
r = 6
c = 4

plt.figure(figsize=(13, 13))

for i, column in enumerate(diabetes_df.columns):
    plt.subplot(r, c, i+1)
    dis_dist = sns.distplot(diabetes_df[column], bins=20)
    dis_dist.set_title(f'{column}')  

plt.tight_layout() 
plt.show()
```
