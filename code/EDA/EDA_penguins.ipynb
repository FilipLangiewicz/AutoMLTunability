# 3. Analysis of values and dependencies in the dataset

```python
# mapping string values to numbers
penguins_df['species'] = penguins_df['species'].map({'Adelie': 1,
                                                     'Gentoo': 0})

penguins_df['island'] = penguins_df['island'].map({'Torgersen': 0,
                                                   'Biscoe': 1,
                                                   'Dream': 2})

penguins_df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

correlation = penguins_df.corr().round(2)
plt.figure(figsize=(12, 8))
sns.heatmap(data=correlation, annot=True, cmap='coolwarm')
```

# 4. Distribution

```python
r = 6
c = 4

plt.figure(figsize=(13, 13))

for i, column in enumerate(penguins_df.columns):
    plt.subplot(r, c, i+1)
    dis_dist = sns.distplot(penguins_df[column], bins=20)
    dis_dist.set_title(f'{column}')  

plt.tight_layout() 
plt.show()
```
