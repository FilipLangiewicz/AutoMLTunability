# Data processing for college dataset 

```python
import pandas as pd

data = pd.read_csv("../../data/college.csv")
```

```python
data.head(10)
```

```python
data['will_go_to_college'] = data['will_go_to_college'].map({
    True : 1,
    False : 0
})
data.head(10)
```

```python
df = data
df.head(10)
```

## Check

```python
import matplotlib.pyplot as plt
import seaborn as sns

import warnings 
warnings.filterwarnings("ignore")

r = 6
c = 4

plt.figure(figsize=(13, 13))

for i, column in enumerate(['will_go_to_college']):
    plt.subplot(r, c, i+1)
    dis_dist = sns.distplot(df[column], bins=20)
    dis_dist.set_title(f'{column}')  

plt.tight_layout() 
plt.show()
```

## Save

```python
df.to_csv("../../data/processed/college.csv")
```
