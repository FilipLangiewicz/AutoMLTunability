# Data processing for credit dataset 

```python
import pandas as pd

data = pd.read_csv("../../data/credit.csv")
```

```python
data.head(10)
```

```python
data['A16'] = data['A16'].map({
    '+' : 1,
    '-' : 0
})
data.head(10)
```

```python
data.info()
```

```python
data[data == "?"] = None
data.isnull().sum()
```

## Check

```python
# mapping string values to numbers
import copy
credit_df = copy.deepcopy(data)
credit_df['A1'] = credit_df['A1'].map({'a': 1, 
                     'b': 0})

credit_df['A2'] = pd.to_numeric(credit_df['A2'], errors='coerce')

credit_df['A4'] = credit_df['A4'].map({'u': 1,
                                       'y': 0,
                                       'l': -1})

credit_df['A5'] = credit_df['A5'].map({'g': 1,
                                        'p': 0,
                                        'gg': -1})

credit_df['A6'] = credit_df['A6'].map({'c': 1,
                                        'd': 0,
                                        'cc': -1,
                                        'i': -2,
                                        'j': -3,
                                        'k': -4,
                                        'm': -5,
                                        'r': -6,
                                        'q': -7,
                                        'w': -8,
                                        'x': -9,
                                        'e': -10,
                                        'aa': -11,
                                        'ff': -12})

credit_df['A7'] = credit_df['A7'].map({'v': 1,
                                        'h': 0,
                                        'bb': -1,
                                        'ff': -2,
                                        'j': -3,
                                        'z': -4,
                                        'dd': -5,
                                        'n': -6,
                                        'o': -7})

credit_df['A9'] = credit_df['A9'].map({'t': 1,
                                       'f': 0})
credit_df['A10'] = credit_df['A10'].map({'t': 1,
                                         'f': 0})

credit_df['A12'] = credit_df['A12'].map({'t': 1,
                                         'f': 0})        

credit_df['A13'] = credit_df['A13'].map({'g': 1,
                                         's': 0, 
                                         'p': -1})

credit_df['A14'] = pd.to_numeric(credit_df['A14'], errors='coerce')


credit_df.head()
```

```python
credit_df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

import warnings 
warnings.filterwarnings("ignore")

r = 6
c = 4

plt.figure(figsize=(13, 13))

for i, column in enumerate(credit_df.columns):
    plt.subplot(r, c, i+1)
    dis_dist = sns.distplot(credit_df[column], bins=20)
    dis_dist.set_title(f'{column}')  

plt.tight_layout() 
plt.show()
```
