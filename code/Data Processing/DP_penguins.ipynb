# Data processing for penguins dataset 

```python
import pandas as pd

data = pd.read_csv("../../data/penguins.csv")
```

```python
data.head(10)
```

```python
data['species'] = data['species'].map({
    'Adelie' : 1,
    'Gentoo' : 0
})
data.head(10)
```

```python
data_columns = data.columns.tolist()
data_columns
```

```python
df = data[(data_columns[1:data.shape[1]] + [data_columns[0]])]
df.head(10)
```

## Check

```python
import matplotlib.pyplot as plt
import seaborn as sns

import warnings 
warnings.filterwarnings("ignore")

r = 6
c = 4

plt.figure(figsize=(13, 13))

for i, column in enumerate(['species']):
    plt.subplot(r, c, i+1)
    dis_dist = sns.distplot(df[column], bins=20)
    dis_dist.set_title(f'{column}')  

plt.tight_layout() 
plt.show()
```

## Save

```python
df.to_csv("../../data/processed/penguins.csv")
```
